### 1.1.3

### 1.1.4
Create the adjacency matrix for the graph, assigning to each character a node identifier equal to the index that the character name has in ascending alphabetical order (remember that the first element of a list in Python has index 0).

```python
#YOUR CODE STARTS HERE#
adj_matrix = []
for from_ in my_set:    #go over all the characters in the estabilished order
  supp = []
  for to_ in my_set:
    supp.append(sum(list(np.multiply(dataset.character_1 == from_,dataset.character_2 == to_))) + 
                sum(list(np.multiply(dataset.character_2 == from_,dataset.character_1 == to_))))
    #multiply boolean arrays ans sum to see if there is a link between two characters  
  adj_matrix.append(supp)
adj_matrix = np.array(adj_matrix)








#YOUR CODE ENDS HERE#
#THIS IS LINE 20#
```

### 1.1.5
Compute the PageRank vector for the obtained graph using a damping factor of 0.8.

```python
#YOUR CODE STARTS HERE#
damping_factor = 0.8
pagerank = sknetwork.ranking.PageRank(damping_factor=damping_factor, solver="piteration", n_iter=1000, tol=10**-6) #set pagerank model
pagerank_vector = pagerank.fit_transform(adj_matrix)  #compute pagerank for the matrix
 













#YOUR CODE ENDS HERE#
#THIS IS LINE 20#
```

Print the name and the PageRank score of the top-10 characters according to the PageRank score of their associated nodes.

```python
#YOUR CODE STARTS HERE#
k = 10
ids = sorted(range(len(pagerank_vector)), reverse = True, key=lambda i: pagerank_vector[i])[0:k]   #sort and take first k
result = pd.DataFrame({'Character': my_set[ids], 'Pagerank score': pagerank_vector[ids]})          #create df
result.index = np.arange(1, len(result)+1)
display(result)


#YOUR CODE ENDS HERE#
#THIS IS LINE 10#
```

### 1.1.6
